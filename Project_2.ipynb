### v. Convert nominal data (categorical) into dummies 

```python
Gender = {'Male': 0, 'Female': 1}
driving_licence_present = {0: 0, 1: 1}
previously_insured = {0: 1, 1: 0}
vehicle_age = {'> 2 Years': 2, '1-2 Year': 1, '< 1 Year': 0}
vehicle_damage = {'Yes': 1, 'No': 0}
```

```python
table1['Gender'] = table1['Gender'].map(Gender)
table1['Driving Licence Present'] = table1['Driving Licence Present'].map(driving_licence_present)
table1['Previously Insured'] = table1['Previously Insured'].map(previously_insured)
table1['Vehicle Age'] = table1['Vehicle Age'].map(vehicle_age)
table1['Vehicle Damage'] = table1['Vehicle Damage'].map(vehicle_damage)
```
