```python
n=1
while n!=4:
    print('\nMain Menu\n\ntype 0 to create dictionary file\n\ntype 1 To Add a new word',
          '\n\ntype 2 To Find the meaning\n\ntype 3 To Update a word\n\ntype 4 To Exit\n')
    
    n=int(input('Enter your choice: '))
    
    if n==4:
        print('\n....exiting app....\n')
        break
    
    elif n==0:
        print('*'*30)
        print('\nCreating file....')
        dict2=dict()
        file=open('words.txt','w')
        json.dump(dict2,file)
        file.close()
        print('\nfile created sucessfully...')
        print('*'*30)
    
    elif n==1:
        print('*'*30)
        file=open('words.txt','r')
        l=json.load(file)
        print('\n')
        w=input('Enter the word: ')
        word=w.title()
        
        if word in l:
            print('\nword already exist..')
            print('*'*30)
            file.close()
            
        else:
            dict1=dict()
            print('\n')
            m=input('Enter the meaning of word: ')
            meaning=m.title()
            dict1[word]=meaning
            l.update(dict1)
            file=open('words.txt','w')
            json.dump(l,file)
            dict1.popitem()
            file.close()
            print('\nword added sucessfully..')
            print('*'*30)
            
            
    elif n==2:
        print('\n')
        print('*'*30)
        w=input('Enter the word: ')
        word=w.title()
        file=open('words.txt','r')
        l=json.load(file)
        
        if word in l:
            print('\nMeaning: ',l[word])
            print('*'*30)
            file.close()
            
            
        else:
            print('\nWord does not exists..')
            print('*'*30)
            file.close()
            
            
    elif n==3:
        print('\n')
        print('*'*30)
        w=input('Enter the word you want to update: ')
        word=w.title()
        file=open('words.txt','r')
        l=json.load(file)
        
        if word in l:
            file.close()
            print('\n')
            m=input('Enter the updated meaning: ')
            meaning=m.title()
            l[word]=meaning
            file=open('words.txt','w')
            json.dump(l,file)
            file.close()
            print('\nSucessfully updated')
            print('*'*30)
        
        else:
            print('\nWord does not exists..')
            print('*'*30)
            file.close()
            
    else:
        print('*'*30)
        print('\ninvalid choice')
        print('*'*30)
```

```python
# importing files
from tkinter import *
import tkinter as tk
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
```

```python
root = Tk()
root.title("Library Management Application")
```

```python
# Methods
def Database():
    global conn,cursor
    conn = sqlite3.connect('simple.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS librarytable (booktitle TEXT, author TEXT, year, isbn )")
    
    
def addbook():
    titlebookname=((booktitle.get()).title())
    titleauthor=((author.get()).title())
    tyear=str(year.get())
    tisbn=str(isbn.get())
    
    if titlebookname=='' or titleauthor=='' or tyear=='' or tisbn=='':
        tk.messagebox.showinfo('Library','Please fill all the details')
        
    else:
        Database()
        cursor.execute("INSERT INTO librarytable (booktitle,author,year,isbn) VALUES(?,?,?,?)",(titlebookname, titleauthor, tyear, tisbn))
        conn.commit()
        booktitle.set("")
        author.set("")
        year.set("")
        isbn.set("")
        cursor.close()
        conn.close()
        txt_result.config(text="Record Inserted ! ",fg ="Green")
        tk.messagebox.showinfo('Library','Values added sucessfully')
        
def display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM librarytable ORDER BY booktitle ASC")
    fetch = cursor.fetchall()
    
    for data in fetch:
        tree.insert('','end',values=[data[0],data[1],data[2],data[3]])
    
    cursor.close()
    conn.close()
    
    txt_result.config(text="Successfully fetched data !",fg='green')
    tk.messagebox.showinfo('Library','Data fetched Successfully')
    
    
def search():
    tree.delete(*tree.get_children())
    count=0
    Database()
    titlebookname=((booktitle.get()).title())
    titleauthor=((author.get()).title())
    tyear=str(year.get())
    tisbn=str(isbn.get())
    cursor.execute("SELECT * FROM librarytable ORDER BY booktitle ASC")
    fetch = cursor.fetchall()
    
    if titlebookname=='' and titleauthor=='' and tyear=='' and tisbn=='':
        tk.messagebox.showinfo('Library','Please fill any 1 detail')
        
    else:
        for data in fetch:
            if titlebookname==data[0] or titleauthor==data[1] or tyear==data[2] or isbn==data[3]:
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
                
                
    
        if count>=1:
            tk.messagebox.showinfo('Library','Search done sucessfully')
        
        else:
            tk.messagebox.showinfo('Library','The book does not Exist')
        
    booktitle.set('')
    author.set('')
    year.set('')
    isbn.set('')
    cursor.close()
    conn.close()
    
    
    
def issue():
    count=0
    Database()
    titlebookname=((booktitle.get()).title())
    titleauthor=((author.get()).title())
    tyear=str(year.get())
    tisbn=str(isbn.get())
    tree.delete(*tree.get_children())
    cursor.execute("SELECT * FROM librarytable ORDER BY booktitle ASC")
    fetch = cursor.fetchall()
    
    if titlebookname=='' and titleauthor=='' and tyear=='' and tisbn=='':
        tk.messagebox.showinfo('Library','Please fill any 1 detail')
        
    else:
        for data in fetch:
            if titlebookname==data[0] or titleauthor==data[1] or tyear==data[2] or isbn==data[3]:
                tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
                count+=1
                    
                
        if count>=1:
            tk.messagebox.showinfo('Library','Book issued')
            
        else:
            tk.messagebox.showinfo('Library','The book does not Exist')
            
    
    booktitle.set('')
    author.set('')
    year.set('')
    isbn.set('')
    cursor.close()
    conn.close()
    
    
    

def delete():
    count=0
    Database()
    titlebookname=((booktitle.get()).title())
    titleauthor=((author.get()).title())
    tyear=str(year.get())
    tisbn=str(isbn.get())
    tree.delete(*tree.get_children())
    cursor.execute("SELECT * FROM librarytable ORDER BY booktitle ASC")
    fetch = cursor.fetchall()
    
    if titlebookname=='' and titleauthor=='' and tyear=='' and tisbn=='':
        tk.messagebox.showinfo('Library','Please fill any 1 detail')
        
    else:
        for data in fetch:
            if titlebookname==data[0] or titleauthor==data[1] or tyear==data[2] or isbn==data[3]:
                cursor.execute("DELETE FROM librarytable where booktitle=? or author=? or year=? or isbn=?",(titlebookname,titleauthor,tyear,tisbn))
                count+=1
                
                
        if count>=1:
            tk.messagebox.showinfo('Library','Book deleted')
        
        else:
            tk.messagebox.showinfo('Library','The book does not exist')
        

    booktitle.set("")
    author.set("")
    year.set("")
    isbn.set("")
    conn.commit()
    cursor.close()
    conn.close()
    
    
    
def exit():
    result = tkMessageBox.askquestion("Do you want to exit ? (y/n)",'do you want to exit',icon="warning")
    if result == 'yes':
        root.destroy()
```

```python
# variable declaration
booktitle=StringVar()
author=StringVar()
year=StringVar()
isbn=StringVar()
```

```python
# Frame:
Top = Frame(root,width=900,height=50,bd=8,relief="raise",bg='blue')
Top.pack(side=TOP)
Left = Frame(root,width=300,height=500,bd=8,relief="raise",bg='red')
Left.pack(side=LEFT)
Right = Frame(root,width=300,height=300,bd=8,relief="raise",bg='yellow')
Right.pack(side=RIGHT)
Forms = Frame(Left,width=300,height=450)
Forms.pack(side=TOP)
Buttons = Frame(Left,width=300,height=100,relief="raise")
Buttons.pack(side=BOTTOM)


# label widget
txt_title = Label(Top,width=900,font=('arial',24),text="...Library Management Application...")
txt_title.pack()

txt_booktitle = Label(Forms,text="Book Title:",font=('arial',16),bd=15)
txt_booktitle.grid(row=0,stick="e")
txt_author = Label(Forms,text="Author:",font=('arial',16),bd=15)
txt_author.grid(row=1,stick="e")
txt_year = Label(Forms,text="Year",font=('arial',16),bd=15)
txt_year.grid(row=2,stick="e")
txt_isbn = Label(Forms,text="ISBN:",font=('arial',16),bd=15)
txt_isbn.grid(row=3,stick="e")

txt_result = Label(Buttons)
txt_result.pack(side=TOP)
```

```python
# Entry Widget:
entry_booktitle = Entry(Forms,textvariable=booktitle, width=30)
entry_booktitle.grid(row=0, column=1)

entry_author = Entry(Forms,textvariable=author, width=30)
entry_author.grid(row=1, column=1)

entry_year = Entry(Forms,textvariable=year, width=30)
entry_year.grid(row=2, column=1)

entry_isbn = Entry(Forms,textvariable=isbn, width=30)
entry_isbn.grid(row=3, column=1)
```

```python
# Button widget

btn_display = Button(Buttons, width=10, text="Display", command=display)
btn_display.pack(side=LEFT)

btn_search = Button(Buttons, width=10, text="Search", command=search)
btn_search.pack(side=LEFT)

btn_add = Button(Buttons, width=10, text="Add", command=addbook)
btn_add.pack(side=LEFT)

btn_issue = Button(Buttons, width=10, text="Issue", command=issue)
btn_issue.pack(side=LEFT)

btn_delete = Button(Buttons, width=10, text="Delete", command=delete)
btn_delete.pack(side=LEFT)

btn_exit = Button(Buttons, width=10, text="Exit", command=exit)
btn_exit.pack(side=LEFT)
```

```python
# List Widget:
scrollbary = Scrollbar(Right, orient=VERTICAL)
scrollbarx = Scrollbar(Right, orient=HORIZONTAL)
col=("booktitle", "author", "year", "isbn")
tree = ttk.Treeview(Right,columns=col,show='headings')

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT, fill=Y)
scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM, fill=X)

tree.heading('booktitle', text="Booktitle", anchor=W)
tree.heading('author', text="Author", anchor=W)
tree.heading('year', text="year", anchor=W)
tree.heading('isbn', text="ISBN", anchor=W)

tree.column('booktitle', stretch=NO, minwidth=0, width=150)
tree.column('author', stretch=NO, minwidth=0, width=150)
tree.column('year', stretch=NO, minwidth=0, width=150)
tree.column('isbn', stretch=NO, minwidth=0, width=150)
tree.pack(side=TOP,fill=BOTH)

tree.pack()
```
